```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Baca gambar
img = cv2.imread('/kaggle/input/contoh-citra-digital/other/default/1/download.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Tampilkan
plt.figure(figsize=(5,5))
plt.imshow(img_rgb)
plt.title("Gambar Asli")
plt.axis("off")
plt.show()
```

Menampilkan Gambar Asli

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(5,5))
plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis("off")
plt.show()

cv2.imwrite("gray_sepatu.jpg", gray)
```

Mengubah Gambar ke Graysclae

```python
resized = cv2.resize(img, (150, 150))

plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.title("Resize ke 150x150")
plt.axis("off")
plt.show()

cv2.imwrite("resized_sepatu.jpg", resized)
```

Gambar ke Resize : 150 x 150

```python
blur = cv2.GaussianBlur(img, (7, 7), 0)

plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.title("Gaussian Blur")
plt.axis("off")
plt.show()

cv2.imwrite("blur_sepatu.jpg", blur)

```

```python
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5,-1],
                           [0, -1, 0]])

sharpen = cv2.filter2D(img, -1, sharpen_kernel)

plt.imshow(cv2.cvtColor(sharpen, cv2.COLOR_BGR2RGB))
plt.title("Sharpen")
plt.axis("off")
plt.show()

cv2.imwrite("sharpen_sepatu.jpg", sharpen)
```

# Filter - Blur & Sharpan

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection (Canny)")
plt.axis("off")
plt.show()

cv2.imwrite("edges_sepatu.jpg", edges)
```

# Edge Detection (Canny)

```python
flip = cv2.flip(img, 1)

plt.imshow(cv2.cvtColor(flip, cv2.COLOR_BGR2RGB))
plt.title("Flip Horizontal")
plt.axis("off")
plt.show()

cv2.imwrite("flip_sepatu.jpg", flip)
```

```python
rotate = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)

plt.imshow(cv2.cvtColor(rotate, cv2.COLOR_BGR2RGB))
plt.title("Rotate 90°")
plt.axis("off")
plt.show()

cv2.imwrite("rotate_sepatu.jpg", rotate)
```
